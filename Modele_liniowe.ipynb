# Regresja liniowa

```python
from sklearn.linear_model import LinearRegression
data = mglearn.datasets.make_wave(n_samples = 60)
```

```python
data
```

```python
X, y = mglearn.datasets.make_wave(n_samples=100)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.title('regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.legend()
plt.plot()
```

```python
y_pred = lr.predict(X)
```

```python
plt.figure(figsize=(8, 6))
plt.title('regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X, y, label='cecha x')
plt.plot(X, y_pred, color='red', label='model')
plt.legend()
plt.plot()
```

# Regresja grzbietowa

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
print(ridge.score(X_train, y_train))
print(ridge.score(X_test, y_test))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print(ridge10.score(X_train, y_train))
print(ridge10.score(X_test, y_test))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print(ridge01.score(X_train, y_train))
print(ridge01.score(X_test, y_test))
```

```python
mglearn.plots.plot_ridge_n_samples()
```

# Model Lasso

```python
from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train, y_train)
print(lasso.score(X_train, y_train))
print(lasso.score(X_test, y_test))
print(np.sum(lasso.coef_ != 0)) #liczba użytych cech
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print(lasso001.score(X_train, y_train))
print(lasso001.score(X_test, y_test))
print(np.sum(lasso001.coef_ != 0)) #liczba uzytych cech
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print(lasso00001.score(X_train, y_train))
print(lasso00001.score(X_test, y_test))
print(np.sum(lasso00001.coef_ != 0)) #liczba uzytych cech
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
X, y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1,2,figsize=(10,3))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
  clf = model.fit(X, y)
  mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
  mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
  ax.set_title("{}".format(clf.__class__.__name__))
  ax.set_xlabel("Cecha 0")
  ax.set_ylabel("Cecha 1")
axes[0].legend()
```

```python
# w przypadku LogisticRegression i LinearSVC parametr kompromisu określa się C, Niski C algorytmy beda próbowały dostować się do wiekszości punktów danych, gdy wyższa C podkresla znaczenie prwidłowego sklasyfikowania każdego pojedynczego punktu danych.
mglearn.plots.plot_linear_svc_regularization()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
  cancer.data, cancer.target, stratify=cancer.target, random_state=42
)
logreg = LogisticRegression().fit(X_train, y_train) #domyslnie jest 1
print(logreg.score(X_train, y_train))
print(logreg.score(X_test, y_test))
```

```python
logreg = LogisticRegression(C=100000).fit(X_train, y_train)
print(logreg.score(X_train, y_train))
print(logreg.score(X_test, y_test))
```
